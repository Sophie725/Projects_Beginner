```python
for stars in range (1, 6):
  # type annotations are for 'functions' only not for 'for' statements
  print(" " * (5 - stars) + "*" * stars)
```

```python
def make_right_pyramid(n) -> None:
  for stars in range(1, n+1):
    for star in range(n - stars):
      print(" ", end = "")
    for sta in range(stars):
      print("*", end = "")

    print()
```

```python
print(make_right_pyramid(6))
```
