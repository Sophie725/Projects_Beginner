```python
def inverted_pyramid(n: int) -> None:
  for stars in range(n, 0, -1): #range(start, stop, step)
    for star in range(stars):
      print("*", end="")
    print()
```

```python
if __name__ == "__main__":
  N = 5
  print(inverted_pyramid(N))
```
