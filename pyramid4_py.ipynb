Inverted, right aligned pyramid

```python
def right_inverted_pyramid():
  for stars in range(1, 6):
    for stars in range(stars):
      print(" ", end = "")
    for sta in range(5 - stars):
      print("*", end = "")

    print()
```

```python
right_inverted_pyramid()
```

Pyramid: # of stars increases by odd numbers

```python
def odd_number_pyramid():
  for spaces in range(1, 10, 2):
    a = int((9-spaces)/2)
    for space in range(a):
      print(" ", end = "")
    for star in range(spaces):
      print("*", end = "")
    print()
```

```python
odd_number_pyramid()
```
