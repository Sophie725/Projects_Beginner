Inverted, right aligned pyramid

```python
def right_inverted_pyramid(n: int):
  for stars in range(1, n+1):
    for stars in range(stars):
      print(" ", end = "")
    for sta in range(n - stars):
      print("*", end = "")

    print()
```

```python
if __name__ == "__main__":
  N = 5
  right_inverted_pyramid(N)
```

Pyramid: # of stars increases by odd numbers

```python
def odd_number_pyramid(n: int):
  num_of_stars = 2 * n
  for spaces in range(1, num_of_stars, 2):
    a = int((num_of_stars - 1 - spaces)/2)
    for space in range(a):
      print(" ", end = "")
    for star in range(spaces):
      print("*", end = "")
    print()
```

```python
if __name__ == "__main__":
  N = 5
  odd_number_pyramid(N)
```
