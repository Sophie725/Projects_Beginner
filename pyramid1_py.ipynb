```python
for stars in range(1,6):
  print("*" * stars)
```

```python
for stars in range(1,6):
  for star in range(stars):
    print("*", end="")
  print()
```

```python
def make_stars(n: int) -> None: #make a function
  for stars in range(1,n): #repeat to make rows from 1st to 5th
    for star in range(stars): #repeat to make star on each row
      print("*", end="")
    print()
```

```python
if __name__ == "__main__":
  N = 6
  make_stars(N)
```
