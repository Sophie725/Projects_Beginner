```python
def middle_inverted_pyr(n: int) -> None:
  for spaces in range(n, 0, -2):
    a = int((n-spaces)/2)
    for space in range(a):
      print(" ", end = "")
    print("*"*spaces)
```

```python
if __name__ == "__main__":
  N: int = 9
  middle_inverted_pyr(N)
```

```python
def vertical_pyr(n: int) -> None:
  for stars in range(1, n):
    star = "* "
    star_count = star * stars
    print(f"{star_count: ^10}")
```

```python
if __name__ == '__main__':
  N: int = 6
  vertical_pyr(N)
```
