```python
def middle_inverted_pyr(n: int) -> None:
  for spaces in range(2 * n - 1, 0, -2):
    a = int((n * 2 -spaces)/2)
    for space in range(a):
      print(" ", end = "")
    print("*"*spaces)
```

```python
if __name__ == "__main__":
  N: int = 10
  middle_inverted_pyr(N)
```

```python
def vertical_pyr(n: int) -> None:
    for i in range(1, n + 1):
        num_spaces = n - i
        for _ in range(num_spaces):
            print(" ", end="")
        for _ in range(i):
            print("* ", end="")
        print()
```

```python
if __name__ == '__main__':
  N: int = 9
  vertical_pyr(N)
```
