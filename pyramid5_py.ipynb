```python
def middle_inverted_pyr():
  for spaces in range(9, 0, -2):
    a = int((9-spaces)/2)
    for space in range(a):
      print(" ", end = "")
    print("*"*spaces)
```

```python
middle_inverted_pyr()
```

```python
def vertical_pyr():
  for stars in range(1, 6):
    star = "* "
    star_count = star * stars
    print(f"{star_count: ^10}")
```

```python
vertical_pyr()
```
