```python
def middle_inverted_pyr(n: int) -> None:
  for spaces in range(2 * n - 1, 0, -2):
    a = int((n * 2 -spaces)/2)
    for space in range(a):
      print(" ", end = "")
    print("*"*spaces)
```

```python
if __name__ == "__main__":
  N: int = 10
  middle_inverted_pyr(N)
```

```python
def vertical_pyr(n: int) -> None:
  for stars in range(1, n + 1):
    star = "* "
    star_count = star * stars
    space_count = 2 * n + 2
    print(f"{star_count: ^{space_count}}")
```

```python
if __name__ == '__main__':
  N: int = 9
  vertical_pyr(N)
```
